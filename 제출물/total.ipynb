```python
import pandas as pd
```

```python
elastic=pd.read_csv('submissison_elastic.csv').rename(columns={'answer':'elastic'})
gradient=pd.read_csv('submissison_gradientboosting.csv').rename(columns={'answer':'gradient'})
lasso=pd.read_csv('submissison_Lasso.csv').rename(columns={'answer':'lasso'})
ridge=pd.read_csv('submissison_ridge.csv').rename(columns={'answer':'ridge'})
xgboost=pd.read_csv('submissison_xgboost.csv').rename(columns={'answer':'xgboost'})
```

```python
elastic
```

```python
elastic = elastic.pivot_table( columns='num_date_time', values='elastic')
gradient = gradient.pivot_table( columns='num_date_time', values='gradient')
lasso = lasso.pivot_table( columns='num_date_time', values='lasso')
ridge = ridge.pivot_table( columns='num_date_time', values='ridge')
xgboost = xgboost.pivot_table( columns='num_date_time', values='xgboost')
```

```python
pd.concat([elastic,gradient,lasso,ridge,xgboost])
```
